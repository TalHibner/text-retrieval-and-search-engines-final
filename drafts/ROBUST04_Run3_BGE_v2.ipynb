# ROBUST04 Run 3 Ensemble: RRF + BGE + SPLADE

## Scenario 3: High Performance - RRF & Discriminative Reranking

### Strategy
- **Stage 1**: Multi-method retrieval (BM25, RM3 tuned)
- **Stage 2**: Reciprocal Rank Fusion (RRF)
- **Stage 3**: 2-model ensemble reranking:
  - BGE (castorini/bge-base-msmarco-10k)
  - SPLADE Retrieval
- **Stage 4**: Final RRF fusion
- **Target MAP**: 0.30-0.35

### Why This Works Well
- **Tuned RM3**: Increased expansion terms (100) for long documents.
- **Discriminative Reranker**: BGE is a dedicated ranker, not a chat model.
- **RRF**: Robust fusion method that doesn't rely on uncalibrated scores.

## 1. Configuration

```python
# Model configuration
USE_BGE = True

print(f"\n🎯 Ensemble Configuration:")
print(f"  • BGE Reranker: {USE_BGE}")
```

## 2. Load Index and Configure Searchers

```python
print("Loading ROBUST04 index...")
index_reader = IndexReader.from_prebuilt_index('robust04')

# Single BM25 searcher - RM3 was hurting performance
searcher = LuceneSearcher.from_prebuilt_index('robust04')
searcher.set_bm25(k1=0.9, b=0.4)

searchers = [searcher]

print("✓ Configured BM25 (k1=0.9, b=0.4)")
print("  RM3 disabled - was hurting baseline")
```

## 3. Load Queries

```python
def load_queries(query_file):
    queries = {}
    with open(query_file, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            parts = line.split(None, 1)
            if len(parts) == 2:
                queries[parts[0]] = parts[1]
    return queries

def load_qrels(qrel_file):
    qrels = defaultdict(dict)
    with open(qrel_file, 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 4:
                qrels[parts[0]][parts[2]] = int(parts[3])
    return qrels

queries = load_queries('queriesROBUST.txt')
qrels = load_qrels('qrels_50_Queries')
train_qids = sorted(qrels.keys())
test_qids = sorted([q for q in queries if q not in qrels])

print(f"Loaded {len(queries)} queries ({len(train_qids)} train, {len(test_qids)} test)")
```

## 4. Helper Functions

```python
def reciprocal_rank_fusion(rank_lists, k=60):
    """
    Combine multiple ranked lists using Reciprocal Rank Fusion.
    rank_lists: List of lists, where each inner list contains (docid, score) tuples 
                sorted by score desc (or just rank order).
    """
    rrf_scores = defaultdict(float)
    
    for rank_list in rank_lists:
        for rank, (docid, _) in enumerate(rank_list):
            rrf_scores[docid] += 1.0 / (k + rank + 1)
            
    return sorted(rrf_scores.items(), key=lambda x: x[1], reverse=True)

def classify_query(query_text):
    wc = len(query_text.split())
    return 'short' if wc <= 3 else 'medium' if wc <= 6 else 'long'

def get_adaptive_weights(query_text):
    qtype = classify_query(query_text)
    if qtype == 'short':
        return [1.5, 1.2, 0.8, 0.5, 1.0]
    elif qtype == 'medium':
        return [1.0, 1.0, 1.2, 1.0, 0.8]
    else:
        return [0.7, 0.8, 1.2, 1.5, 0.9]
```

## 5. Multi-Method Retrieval + Fusion

```python
def multi_method_fusion(query_text, k=60):
    """
    Pure BM25 retrieval (no fusion needed with single method).
    """
    hits = searchers[0].search(query_text, k=1000)
    return [(h.docid, h.score) for h in hits]
```

## 6. Load Reranking Models

```python
# BGE Reranker - Modern Cross-Encoder
bge_model, bge_tokenizer = None, None
if USE_BGE:
    print("Loading BGE Reranker (BAAI/bge-reranker-v2-m3)...")
    
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    from transformers import AutoTokenizer, AutoModelForSequenceClassification
    
    try:
        model_name = "BAAI/bge-reranker-v2-m3"
        bge_tokenizer = AutoTokenizer.from_pretrained(model_name)
        bge_model = AutoModelForSequenceClassification.from_pretrained(model_name).to(device)
        bge_model.eval()
        print(f"  ✓ BGE Reranker loaded successfully")
    except Exception as e:
        print(f"  BGE load failed: {e}")
        USE_BGE = False
```

## 7. Individual Reranker Functions

```python
def rerank_bge(query, doc_ids, batch_size=16):
    """BGE Reranker scoring (cross-encoder).
    
    Returns dict of {doc_id: relevance_score}
    Higher score = more relevant
    """
    if not USE_BGE or bge_model is None:
        return None
    
    doc_texts, valid_ids = [], []
    for did in doc_ids:
        try:
            doc = index_reader.doc(did)
            if doc:
                raw = doc.raw()
                if raw:
                    doc_texts.append(raw[:1500])
                    valid_ids.append(did)
        except:
            pass
    
    if not doc_texts:
        return None
    
    # BGE uses query-document pairs
    pairs = [[query, doc] for doc in doc_texts]
    
    all_scores = []
    for i in range(0, len(pairs), batch_size):
        batch = pairs[i:i+batch_size]
        try:
            inputs = bge_tokenizer(
                batch, 
                return_tensors="pt", 
                padding=True, 
                truncation=True, 
                max_length=512
            ).to(device)
            
            with torch.no_grad():
                outputs = bge_model(**inputs)
                scores = outputs.logits.squeeze(-1).cpu().tolist()
                if isinstance(scores, float):
                    scores = [scores]
                all_scores.extend(scores)
        except Exception as e:
            print(f"  Batch error: {e}")
            all_scores.extend([0.0] * len(batch))
    
    return dict(zip(valid_ids, all_scores))
```

## 8. Ensemble Fusion

```python
def ensemble_fusion(score_dicts):
    """
    Fuse multiple reranker score dicts using RRF.
    """
    rank_lists = []
    for scores in score_dicts:
        ranked = sorted(scores.items(), key=lambda x: x[1], reverse=True)
        rank_lists.append(ranked)
        
    return dict(reciprocal_rank_fusion(rank_lists))
```

## 10. Generate Run 3 Ultimate

```python
def ultimate_pipeline(query, rerank_depth=100, k_rrf=60):
    """
    Proper RRF fusion between BM25 and BGE rankings.
    
    Based on Cormack et al. SIGIR 2009:
    RRF_score(d) = sum(1/(k + rank_i(d))) for each ranking i
    """
    # Stage 1: Get BM25 ranking
    bm25_results = multi_method_fusion(query)
    bm25_ranking = {docid: rank for rank, (docid, _) in enumerate(bm25_results, 1)}
    
    # Stage 2: Get BGE ranking for top docs
    if USE_BGE and bge_model is not None:
        top_k_docids = [docid for docid, _ in bm25_results[:rerank_depth]]
        bge_scores = rerank_bge(query, top_k_docids)
        
        if bge_scores and len(bge_scores) > 0:
            # Convert BGE scores to ranking
            sorted_by_bge = sorted(bge_scores.items(), key=lambda x: x[1], reverse=True)
            bge_ranking = {docid: rank for rank, (docid, _) in enumerate(sorted_by_bge, 1)}
            
            # Stage 3: RRF Fusion
            rrf_scores = {}
            all_docids = set(bm25_ranking.keys())
            
            for docid in all_docids:
                bm25_rank = bm25_ranking.get(docid, 1000)
                
                if docid in bge_ranking:
                    bge_rank = bge_ranking[docid]
                    rrf_score = 1.0/(k_rrf + bm25_rank) + 1.0/(k_rrf + bge_rank)
                else:
                    rrf_score = 1.0/(k_rrf + bm25_rank)
                
                rrf_scores[docid] = rrf_score
            
            final_ranking = sorted(rrf_scores.items(), key=lambda x: x[1], reverse=True)
        else:
            final_ranking = [(docid, 1.0/(k_rrf + rank)) for rank, (docid, _) in enumerate(bm25_results, 1)]
    else:
        final_ranking = [(docid, 1.0/(k_rrf + rank)) for rank, (docid, _) in enumerate(bm25_results, 1)]
    
    results = []
    for rank, (docid, score) in enumerate(final_ranking[:1000], 1):
        results.append((str(docid), float(score), int(rank)))
    
    return results
```
